```r
library(cluster)
library(ggplot2)
```

```r
set.seed(20)
```

```r
irisCluster <- kmeans (iris [,3:4],3,nstart =20)
```

```r
irisCluster
```

```r
irisCluster$cluster <- as.factor(irisCluster$cluster)
```

```r
ggplot(iris,aes(Petal.Length,Petal.Width,color = irisCluster$cluster)) + geom_point()
```
